```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
sample = load_breast_cancer()
```

```python
print(sample.DESCR)
```

```python
features=sample.data
target=sample.target
#Print their shape

print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.2)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
```

```python
#hidden layer 1
model.add(Dense(40,activation="relu",input_dim=30))#Input dim = 30 as there are 30 features in the datset
#hidden layer 2
model.add(Dense(64,activation="relu"))
#OP Layer
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train,Y_train,epochs=500,batch_size=32)
```

```python
score=model.evaluate(X_test,Y_test)
print(score)
```

```python
predictions = model.predict(X_test)
label=[]
i=0
for pred in predictions:
  i+=1
  if pred > 0.5:
    print("Case",i,": Malignant , Prediction : ",pred)
  else:
    print("Case",i,": Benign    , Prediction : ",pred)
```
